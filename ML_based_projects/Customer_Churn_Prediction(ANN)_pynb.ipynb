```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df.drop(columns=['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#one hot encoding
df = pd.get_dummies(df,columns=['Geography','Gender'],drop_first=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['Exited'],axis=1),df['Exited'],test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(32,activation='relu',input_dim=11))#input layer
model.add(Dense(16,activation='relu'))#hidden layer
model.add(Dense(1,activation='sigmoid'))#output layers
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=10,batch_size=8)
```

```python
model.layers[0].get_weights()
```

```python
y_log=model.predict(x_test)
```

```python
y_pred=np.where(y_log>0.5,1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
